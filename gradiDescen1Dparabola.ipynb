### Project 1: 1D Gradient Descent on a Parabola
#### Objective: Implement gradient descent to find the minimum of a simple 1D function.

##### Step 1: Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Define the function and its derivative

```python
def f(x):
    return x**2

def df(x):
    return 2*x

# f(x) represents our function f(x)=x^2.
# df(x) is its derivative f′(x)=2x.
# The derivative tells us the slope of the curve at any point — this helps us know which direction to move (uphill or downhill).
```

##### Step 3: Initialize the parameters

```python
x = 5             # starting point
learning_rate = 0.14
epochs = 20

# x = 5: Starting point on the parabola
# learning_rate = 0.1: Controls how big each step is when moving downhill.
# epochs = 20: Number of times we update x
```

##### Step 4: Gradient Descent loop

```python
x_history = [x]   # store x values to visualize later

for i in range(epochs):
    grad = df(x)                              # find slope at current point
    x = x - learning_rate * grad              # update x using gradient formula
    x_history.append(x)                       # store new x
    print(f"Epoch {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
# If the slope (grad) is positive, we move left (decrease x).
# If the slope is negative, we move right (increase x).
# Each loop moves x closer to where the slope = 0 (the minimum point).
# x_history saves all the intermediate x-values to show the “path” towards the minimum in the graph.
```

##### Step 5: Plotting and visualization

```python
x_vals = np.linspace(-6, 6, 100)
y_vals = f(x_vals)

# Creates smooth curve points for f(x)=x^2 between -6 and 6

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label="f(x) = x²")
plt.scatter(x_history, [f(i) for i in x_history], color='red', label="Descent Path")
plt.title("1D Gradient Descent on a Parabola")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
