### Project 2: Gradient Descent with Multiple Features
#### Objective: Extend the linear regression project to handle two features (e.g., y = 2x1 + 3x2 + 5 + noise). 
#### Technique: Use a NumPy array for the parameters [m1, m2, b] and perform the matrix multiplication @ for the predictions. Calculate the gradient vector with respect to each parameter.

##### Step 1: Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Generate data

```python
np.random.seed(42)
n_samples = 100

X1 = np.random.rand(n_samples) * 10
X2 = np.random.rand(n_samples) * 10
noise = np.random.randn(n_samples) * 2
Y = 2 * X1 + 3 * X2 + 5 + noise # true relationship with noise y = 2x1 + 3x2 + 5 + noise

# X1 and X2: 100 random numbers between 0 and 10 (our two features).
# noise: random small variations (makes data more realistic, not perfectly linear)

# Combine features
X = np.c_[X1, X2, np.ones(n_samples)]

# np.c_ means “concatenate columns.”
# This joins X1, X2, and a column of 1’s (for the bias b) into one matrix
# The shape is (100, 3).

# Initialize
params = np.random.randn(3)
learning_rate = 0.0005
epochs = 500

# params = [m1, m2, b] initialized randomly.
# learning_rate: how big each gradient descent step is.
# epochs: how many times it’ll update the parameters.

# Store loss over time
loss_history = []
```

##### Step 3: Gradient Descent

```python

for _ in range(epochs):
    y_pred = X @ params

# @ = matrix multiplication.
# This gives predicted y-values:
# ypred= m1x1 + m2x2 + b

# Compute error
    error = y_pred - Y
# Difference between predicted and true y-values.
# If error is large, the model is still far from the true line

# Mean Squared Error
    loss = np.mean(error ** 2)
    loss_history.append(loss)

# Mean Squared Error (MSE): average of squared differences between predicted and actual y.
# Smaller loss means better fit

# Compute gradient for [m1, m2, b]    
    gradients = (2 / n_samples) * (X.T @ error)
# This computes the gradient vector (partial derivatives) for each parameter:
# How much to adjust m₁
# How much to adjust m₂
# How much to adjust b
# X.T means transpose of X (to align dimensions for matrix math

#  Update Parameters
    params = params - learning_rate * gradients

# Print results
print(f"m1 = {params[0]:.4f}, m2 = {params[1]:.4f}, b = {params[2]:.4f}")
print(f"Final Loss = {loss_history[-1]:.4f}")

# Plot
plt.plot(loss_history, color='purple')
plt.title("Loss Convergence for Multiple Feature Gradient Descent")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.show()
```
