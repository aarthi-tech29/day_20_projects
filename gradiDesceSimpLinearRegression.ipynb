### Project 2: Gradient Descent for Simple Linear Regression 
#### Objective: Implement gradient descent from scratch for a single-feature linear regression model. 

##### Step 1: Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Generate Data

```python
np.random.seed(42)
X = np.random.rand(50) * 10
noise = np.random.randn(50) * 2
Y = 2 * X + 5 + noise
# np.random.seed(42) → Fixes the random values so the output is the same every time run it.
# np.random.rand(50) * 10 → Generates 50 random x-values between 0 and 10.
# np.random.randn(50) * 2 → Creates random noise (some random variation).
# Y = 2X + 5 + noise → The true relationship is y=2x+5, but add noise to make it look like real-world data (not perfectly straight)
```

##### Step 3: Initialize Parameters

```python
# np.random.seed(42)
m = np.random.randn()
b = np.random.randn()
print(m,b)

# m → slope
# b → intercept
# The algorithm will adjust these values step-by-step to find the best-fitting line
```

##### Step 4: Set Hyperparameters

```python
learning_rate = 0.005
epochs = 300

# learning_rate: How big each step should be during updates (too high = overshoot, too low = slow).
# epochs: How many times it’ll loop through all the data.
```

##### Step 5: Gradient Descent Loop

```python
loss_history = []

# store the loss (error) after each step to see how it changes

# Inside the Loop
for i in range(epochs):
  Y_pred = m * X + b
  error = Y_pred - Y
  loss = np.mean(error ** 2)
  loss_history.append(loss)

# Y_pred → Model’s predicted values using current m and b.
# error → Difference between predicted and actual values.
# loss → Mean Squared Error (average of squared differences).
# Store the loss to plot later.

# Compute the Gradients
  dm = (2/len(X)) * np.sum(X * error)
  db = (2/len(X)) * np.sum(error)

# These formulas come from partial derivatives of the MSE with respect to m and b.
# It tell us which direction to move m and b to make the error smaller.

# Update Parameters
  m -= learning_rate * dm
  b -= learning_rate * db
# Subtract the gradient times the learning rate → moves “downhill” in the loss curve.
# After every loop (epoch), m and b get closer to the best possible values.

# print every 10th epoch
  if i % 10 == 0:
    print(f"Epoch {i}: m = {m:.4f}, b = {b:.4f}, Loss = {loss:.4f}")

    # Every 10 epochs, it print:
    # Current slope m
    # Current intercept b
    # Current loss value

```

##### Step 6: Plot the Results

```python
# Plot 1 – Data and Line
plt.subplot(1, 2, 1)
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, m * X + b, color='red', label='Best Fit Line')
# Blue = Original data points
# Red = Final best-fit line after learning

# Plot 2 – Loss Over Time
plt.subplot(1, 2, 2)
plt.plot(loss_history, color='green', linewidth=2)
plt.title("Loss over Epochs")
# This shows how the loss (error) goes down as the model learns.
# A smooth downward curve means the gradient descent worked properly.

# Display Final Model
print(f"\nFinal model: y = {m:.2f}x + {b:.2f}")
```
