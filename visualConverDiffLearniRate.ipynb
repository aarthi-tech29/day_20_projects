### Project 1: Visualizing Convergence with Different Learning Rates
#### Objective: Run the 1D gradient descent from the mini-project with three different learning rates (e.g., 0.01, 0.1, 1.0) and plot the paths on the same graph to demonstrate the effect of learning rate. 
#### Technique: plt.plot() for each path, showing how a high learning rate can overshoot and a low one can take too long.

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Define function and its derivative

```python
def f(x):
    return x**2

def df(x):
    return 2*x
# f(x)=x^2 → the function we’re minimizing (a simple parabola).
# f′(x)=2x → tells us the slope at each x (the gradient).
# The slope helps us know which direction to move.
```

##### Step 3: Define different learning rates

```python
learning_rates = [0.01, 0.1, 1.0]
# 0.01 → too small (slow)
# 0.1 → just right (good)
# 1.0 → too large (unstable)
```

##### Step 4: Prepare to store paths

```python
paths = {}
```

##### Step 5: Run Gradient Descent for Each Learning Rate

```python
for lr in learning_rates:
    x = 8.0                # starting point
    path = [x]             # store first value
    epochs = 25            # number of steps to run
    
    for _ in range(epochs):
        x = x - lr * df(x)  # update rule
        path.append(x)      # store new x each time
    
    paths[lr] = path        # store the whole path for that learning rate

# df(x) gives the slope at the current point.
# x - lr * df(x) moves x against the slope (downhill).
# A small learning rate → takes tiny steps toward the minimum.
# A large learning rate → jumps too far, might overshoot.
```

##### Step 6: Plot the Parabola

```python
x_vals = np.linspace(-10, 10, 100)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, 'k--', label='f(x) = x²')

# enumerate()- Gives both index and value while looping
# i - Index (0, 1, 2, …)
# lr - Value (learning rate)
# colors[i] - Picks the color for each learning rate
# This creates the curved graph of the parabola itself (-- means dashed black line).

# Plot the paths
colors = ['b', 'g', 'r']
for i, lr in enumerate(learning_rates):
    plt.plot(paths[lr], f(np.array(paths[lr])), 'o-', color=colors[i],
             label=f'Learning Rate = {lr}')

# Each learning rate’s path is plotted:
# 'o-' means points connected with a line.
# Blue → 0.01
# Green → 0.1
# Red → 1.0
# This shows how each rate moves along the parabola toward (or away from) 0.

# Add title, labels, and grid
plt.title('Effect of Learning Rate on Gradient Descent (f(x) = x²)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# 0.01 - blue - Very slow, tiny steps down the parabola - → slow convergence
# 0.01 - green - Smooth and steady convergence to 0 - → perfect balance
# 1.0 - red - Overshoots, oscillates, may diverge - → it jumps over the minimum and oscillates.
```
